```python
from imutils import paths
import os

imagePaths = list(paths.list_images("D:\\train\\train"))

for imagePath in imagePaths:
    print(imagePath)
    name = imagePath.split(os.path.sep)[-2]
    print(name)
```

```python
import random
import cv2
import numpy as np
random.seed(21)
random.shuffle(imagePaths)
labels = []
data = []

for imagePath in imagePaths:
    image = cv2.imread(imagePath)
    image = cv2.resize(image, (64, 64))
    data.append(image)
    name = imagePath.split(os.path.sep)[-2]
    
    if  name == "Bicycle":
        label = np.array([1,0, 0, 0])
    elif name == "Bus":
        label = np.array([0,1, 0, 0])
    elif name == "Car":
        label = np.array([0,0, 1, 0])
    elif name == "Motorcycle":
        label = np.array([0,0, 0, 1])
       
    labels.append(label)
```

```python
from numpy import save
test_data = np.array(data, dtype="float")
test_labels = np.array(labels, dtype="float")
print(test_data.shape)
```

```python
save("yadav_data2.npy",test_data)
save("yadav_target2.npy",test_labels)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras

model = keras.models.load_model("test_anumeration.h5")
```

```python
pre = model.predict(test_data)
print(pre)
predicted = np.argmax(pre,axis=1)
print(predicted)
```

```python
pred = model.predict_proba(test_data)
print(pred)
```

```python
actual = np.argmax(test_labels,axis=1)
print(actual)
```

```python
# confusion matrix in sklearn
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
# confusion matrix
matrix = confusion_matrix(actual,predicted, labels=[0,1,2,3])
print('Confusion matrix : \n',matrix)
```

```python
# outcome values order in sklearn
tp, fn, fp, tn = confusion_matrix(actual,predicted,labels=[0,1,2,3])
print("Outcome values : \n", tp, fn, fp, tn)
```

```python
report = classification_report(actual,predicted,labels=[0,1,2,3])
print("Classification Reprot : \n", report)
```
